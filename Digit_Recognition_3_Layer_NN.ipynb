### Dimensions of parameters W1, b1, Z1,  A1......W3, b3, Z3, A3 = y_hat

### Helper Functions

```python
def initialize_parameters(n_x, n_h1,n_h2, n_y):
    W1 = np.random.randn(n_h1, n_x)*0.01
    W2 = np.random.randn(n_h2, n_h1)*0.01
    W3 = np.random.randn(n_y, n_h2)*0.01
    b1 = np.zeros((n_h1, 1))
    b2 = np.zeros((n_h2, 1))
    b3 = np.zeros((n_y, 1))
    return W1, W2, W3, b1, b2, b3
    
def sigmoid(z):
    return 1/(1+np.exp(-z))

def relu(z):
    return np.maximum(0, z)

def softmax(z):
#     ez = np.exp(z)
#     ez_sum = np.sum(ez, axis=0, keepdims=True)
#     return ez/ez_sum
    return np.exp(z)/sum(np.exp(z))

def compute_z(W, X, b):
    return np.dot(W, X) + b

def compute_activation(z, activation):
    if(activation == "sigmoid"):
        return sigmoid(z)
    elif(activation == "relu"):
        return relu(z)
    elif(activation == "softmax"):
        return softmax(z)

def sigmoid_prime(z):
    return (sigmoid(z) * (1-sigmoid(z)))

def relu_prime(z):
    return z > 0

def one_hot(Y):
    one_hot_Y = np.zeros((Y.size, Y.max() + 1))
    one_hot_Y[np.arange(Y.size), Y] = 1
    one_hot_Y = one_hot_Y.T
    return one_hot_Y

def get_predictions(A3):
    return np.argmax(A3, 0)

def get_accuracy(predictions, Y):
    print(predictions, Y)
    return np.sum(predictions == Y) / Y.size
    
```

```python
def forward_prop(X, parameters):
    
    W1 = parameters["W1"]
    W2 = parameters["W2"]
    W3 = parameters["W3"]
    b1 = parameters["b1"]
    b2 = parameters["b2"]
    b3 = parameters["b3"]
    
    
    Z1 = compute_z(W1, X, b1)
    A1 = compute_activation(Z1, "relu")
    
    Z2 = compute_z(W2, A1, b2)
    A2 = compute_activation(Z2, "sigmoid")
    
    Z3 = compute_z(W3, A2, b3)
    A3 = compute_activation(Z3, "softmax")
       
    cache = {
        "Z1" : Z1,
        "Z2" : Z2,
        "Z3" : Z3,
        "A1" : A1,
        "A2" : A2,
        "A3" : A3
    }
    return cache

def backward_prop(parameters, cache, Y, X):
    
        one_hot_Y = one_hot(Y)
        
        W1 = parameters["W1"]
        W2 = parameters["W2"]
        W3 = parameters["W3"]
        b1 = parameters["b1"]
        b2 = parameters["b2"]
        b3 = parameters["b3"]
        
        Z1 = cache["Z1"]
        Z2 = cache["Z2"]
        Z3 = cache["Z3"]
        A1 = cache["A1"]
        A2 = cache["A2"]
        A3 = cache["A3"]
        
        m = W1.shape[1]
        
        dZ3 = A3 - one_hot_Y
        dW3 = (1/m)*np.dot(dZ3, A2.T)
        db3 = (1/m)*np.sum(dZ3, axis = 1, keepdims=True)
        
        dZ2 = np.dot(W3.T , dZ3) * sigmoid_prime(Z2)
        dW2 = (1/m)*np.dot(dZ2, A1.T)
        db2 = (1/m)*np.sum(dZ2, axis = 1, keepdims=True)
        
        dZ1 = np.dot(W2.T , dZ2) * relu_prime(Z1)
        dW1 = (1/m)*np.dot(dZ1, X.T)
        db1 = (1/m)*np.sum(dZ1, axis = 1, keepdims=True)

        grads = {
              "dW1" : dW1, 
              "dW2" : dW2,
              "dW3" : dW3,
              "db1" : db1,
              "db2" : db2,
              "db3" : db3
            }
        
        return grads
       
```

```python
def update_parameters(grads, parameters, learning_rate):
        W1 = parameters["W1"]
        W2 = parameters["W2"]
        W3 = parameters["W3"]
        b1 = parameters["b1"]
        b2 = parameters["b2"]
        b3 = parameters["b3"]
        
        dW1 = grads["dW1"]
        dW2 = grads["dW2"]
        dW3 = grads["dW3"]
        db1 = grads["db1"]
        db2 = grads["db2"]
        db3 = grads["db3"]
        
        W1 = W1 - learning_rate * dW1
        W2 = W2 - learning_rate * dW2
        W3 = W3 - learning_rate * dW3
        b1 = b1 - learning_rate * db1
        b2 = b2 - learning_rate * db2
        b3 = b3 - learning_rate * db3
        
        parameters = {
              "W1" : W1, 
              "W2" : W2,
              "W3" : W3,
              "b1" : b1,
              "b2" : b2,
              "b3" : b3
            }
    
        return parameters
```

```python
def gradient_descent(X, Y, learning_rate, iterations):
    
    m, n = X.shape
    n_x = m
    n_h1 = 10
    n_h2 = 6
    n_y = 10
    W1, W2, W3, b1, b2, b3 = initialize_parameters(n_x, n_h1, n_h2, n_y)
    
    parameters = {
              "W1" : W1, 
              "W2" : W2,
              "W3" : W3,
              "b1" : b1,
              "b2" : b2,
              "b3" : b3
            }
    
    for i in range(iterations):
        cache = forward_prop(X, parameters)
        grads = backward_prop(parameters, cache, Y, X)
        parameters = update_parameters(grads, parameters, learning_rate)
        if(i%10 == 0):
            print("Iteration: ", i)
            predictions = get_predictions(cache["A3"])
            print(get_accuracy(predictions, Y))
    return parameters
        
```

```python
parameters = gradient_descent(x_train, y_train, learning_rate=0.005, iterations=700)
```

```python
def make_predictions(X, parameters):
    cache = forward_prop(X, parameters)
    predictions = get_predictions(cache["A3"])
    return predictions

def test_prediction(index, parameters):
    current_image = x_train[:, index, None]
    prediction = make_predictions(x_train[:, index, None], parameters)
    label = y_train[index]
    print("Prediction: ", prediction)
    print("Label: ", label)
    
    current_image = current_image.reshape((28, 28)) * 255
    plt.gray()
    plt.imshow(current_image, interpolation='nearest')
    plt.show()
```
